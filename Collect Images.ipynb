## 1. Make Imports

```python
# Import opencv
import cv2 

# Import uuid
import uuid

# Import Operating System
import os

# Import time
import time

# Import text to speech
import pyttsx3
```

## 2. Setup Images Path

```python
IMAGES_PATH = 'Tensorflow/workspace/images/collectedimages'
engine = pyttsx3.init()
```

## 3. Setup Labels and set number of images per symbol

```python
labels = ['Hello', 'VJIT', 'ThankYou', 'ILoveYou', 'More', 'Pray', 'BestOfLuck', 'No', 'Remember',
         'Nice', 'Home']
number_imgs = 25
```

## 4. Iterate and capture images for each label

```python

for label in labels:
    !mkdir {'Tensorflow\workspace\images\collectedimages\\'+label}
    cap = cv2.VideoCapture(0)
    print('Collecting images for {}'.format(label))
    engine.say('Collecting images for {}'.format(label))
    engine.runAndWait()
    time.sleep(5)
    for imgnum in range(number_imgs):
        if (imgnum%10 == 0) and (imgnum != 0):
            engine.say("Change position")
            engine.runAndWait()
            time.sleep(2)
        print('Collecting image {}'.format(imgnum))
        ret, frame = cap.read()
        # print(ret, frame)
        imgname = os.path.join(IMAGES_PATH,label,label+'.'+'{}.jpg'.format(str(uuid.uuid1())))
        # print(imgname)
        cv2.imwrite(imgname, frame)
        cv2.imshow('frame', frame)
#         engine.say('Image Saved')
#         engine.runAndWait()
        time.sleep(1)
        
        if cv2.waitKey(1) & 0xFF == ord('w'):
            engine.say("Waiting")
            engine.runAndWait()
            break

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

cap.release()

```
